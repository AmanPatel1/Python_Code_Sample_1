```
import pandas as pd
import random
import folium
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSUwMrnp914KqqSw0G7T-L8cJM9qNgXHpO6K-NdM-vLdPBkNfUBZoNcKvTGJBjI7u8Wxxp_uWsH3CwH/pub?gid=1766707236&single=true&output=csv')

m = folium.Map(location=[df['Lat'].mean(), df['Long'].mean()], tiles=None, zoom_start=5)
folium.TileLayer(tiles='openstreetmap', show=True, control=False, min_zoom=3).add_to(m)

list = ['Confederate', 'Union', 'Inconclusive']
c_list = ['gray', 'blue', 'red', 'orange']


for group in list: #OUTER for-loop:compacts the program by filtering the dataframe by each of the three outcomes
  group_df = df[df['Who Won'] == group]
  f_group = folium.FeatureGroup(name = group, show=True)
  color = random.choice(c_list)
  c_list.remove(color)
  for i in range(0,len(group_df)): #INNER for-loop: Plots the points for each different outcome on the map
      folium.Circle(
      location = [group_df.iloc[i]['Lat'], group_df.iloc[i]['Long']],
      radius = float(group_df.iloc[i]['Total Casualties'])*2.5,
      popup = "Battle: " + str(group_df.iloc[i]['Battle']) + "\n" + "State: " + str(group_df.iloc[i]['State']) + "\n" \
      + "Total Number of Casualties: " + str(group_df.iloc[i]['Total Casualties']),
      color = color,
      stroke = False,
      opacity = 100,
      fill_color = color
  ).add_to(f_group)
  f_group.add_to(m)
folium.LayerControl(collapsed=False).add_to(m)
m
```
